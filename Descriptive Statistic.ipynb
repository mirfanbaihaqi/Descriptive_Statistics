# Correlation

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')

plt.scatter(x = data['average_montly_hours'], y = data['satisfaction_level'], c = data['promotion_last_5years'])
plt.xlabel('aaverage_montly_hours')
plt.ylabel('satisfaction_level')
plt.show()
```

```python
np.corrcoef(data['average_montly_hours'],data['satisfaction_level'])
```

```python
data.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(), annot=True)
plt.show()
```
